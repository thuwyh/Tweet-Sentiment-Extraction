```python
import pandas as pd
import numpy as np

%matplotlib inline
```

```python
train = pd.read_csv('../input/tweet-sentiment-extraction/train_folds.csv')
```

```python
# train = pd.read_csv('../input/tweet-sentiment-extraction/tweet_dataset.csv')
```

```python
train.dropna(subset=['selected_text'], inplace=True)
train.drop_duplicates(subset=['text'], inplace=True)
```

```python
train['text'] = train['text'].apply(lambda x: ' '.join(x.strip().split()))
train['selected_text'] = train['selected_text'].apply(lambda x: ' '.join(x.strip().split()))
```

```python
def get_pos(x, y):
    return x.find(y)
```

```python
train['start_pos'] = train.apply(lambda x: get_pos(x['text'], x['selected_text']), axis=1)
```

```python
def broken(x, y):
    if y>0 and x[y-1] not in [' ','.',',','!','?']:
        return True
    return False
```

```python
train['broken'] = train.apply(lambda x: broken(x['text'], x['start_pos']), axis=1)
```

```python
train[train.broken].sample(n=5)
```

```python
train['broken_start'] = train['selected_text'].str[0]
```

```python
print(train[train.broken]['broken_start'].value_counts().head(20))
```

```python
def find_last_char(x, y):
    if y>1 and x[y-1]==' ':
        return x[y-2]
    return np.nan
```

```python
train['prev_word_last_char'] = train.apply(lambda x: find_last_char(x['text'], x['start_pos']), axis=1)
```

```python
print(train[~train.broken]['prev_word_last_char'].value_counts().head(20))
```

```python
train[(~train.broken)&(train.prev_word_last_char=='s')].sample(n=5)
```

```python
def get_prev_word(x, y):
    if y>1 and x[y-1]==' ':
        s = y-2
        while s>=0 and x[s]!=' ':
            s = s-1
        return x[s+1:y-1]
    if y>1 and x[y-1]!=' ':
        s = y-1
        while s>=0 and x[s]!=' ':
            s = s-1
        return x[s+1:y+1]
    return np.nan
```

```python
train['prev_word'] = train.apply(lambda x: get_prev_word(x['text'], x['start_pos']), axis=1)
```
